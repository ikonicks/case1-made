Системы обнаружения вторжений (IDS) анализируют сетевой трафик, выявляя аномальные и
подозрительные активности. Для их обучения используются наборы данных, содержащие
примеры нормального и вредоносного трафика. Одним из таких датасетов является NSL-KDD,
представляющий собой улучшенную версию KDD'99, с устраненными дублирующимися
записями и улучшенным балансом классов.
Анализ данных включает изучение их структуры, распределения признаков и выявление
закономерностей. Визуализация помогает лучше понять данные и обнаружить возможные
аномалии. Предобработка включает обработку пропущенных значений, кодирование
категориальных признаков и нормализацию, что повышает точность работы моделей
машинного обучения.

DoS (Denial of Service)  
Атака на отказ в обслуживании. Ее цель — сделать ресурс (например, веб-сайт или сервер) недоступным для пользователей. Это достигается за счёт перегрузки системы запросами.

R2L (Remote to Local)  
Атаки, при которых злоумышленник пытается получить несанкционированный доступ к удалённой системе и возможность выполнять команды с правами настоящего пользователя.

U2R (User to Root)  
Этот тип атак направлен на получение привилегий суперпользователя (администратора) на уже скомпрометированной системе. Злоумышленник, обладая ограниченными правами, пытается повысить их до уровня администратора.

Probe (Сканирование)  
Это предварительная атака, при которой злоумышленник исследует сеть, чтобы выявить активные устройства, открытые порты и потенциально уязвимые службы. Цель — собрать информацию для последующих атак.

Ботнеты  
Это сеть заражённых устройств (например, компьютеров или IoT-устройств), которые злоумышленник контролирует дистанционно. Ботнеты часто используются для проведения масштабных атак, таких как DoS, спам-рассылки или кража данных.

Атаки с использованием эксплойтов  
Злоумышленники используют специальные программы или коды (эксплойты), чтобы воспользоваться уязвимостями в программном обеспечении или системе. Это позволяет получить несанкционированный доступ или выполнить нежелательные действия на атакуемой системе.

Вредоносное ПО (Malware)  
Под этим термином понимаются различные виды вредоносных программ, включая вирусы, трояны, шпионское ПО и ransomware, которые заражают устройства, чтобы нанести вред, похитить данные или выполнить другие нежелательные действия.

Теперь о наборах данных, которые мы будем использовать в работе. NSL-KDD, CICIDS2017 и
UNSW-NB15 — это популярные наборы данных, используемые для исследования и разработки
систем обнаружения вторжений. Каждый из них содержит информацию о сетевых соединениях
и помогает анализировать поведение нормального и вредоносного трафика.
NSL-KDD является улучшенной версией KDD'99, созданной для устранения его недостатков,
таких как дублирование записей и дисбаланс классов. Данные включают сетевые соединения с
различными атрибутами, такими как продолжительность сессии, количество переданных
байтов и использованный протокол. Метки классов указывают, является ли трафик нормальным
или относится к одному из видов атак, включая DoS, R2L, U2R и Probe.
CICIDS2017 он включает реальные сетевые атаки, такие как ботнеты, атаки с использованием
эксплойтов и вредоносного ПО, а также трафик обычных пользователей. В этом датасете
представлены детализированные сетевые характеристики и метки классов, что делает его
полезным для обучения моделей обнаружения вторжений.
UNSW-NB15 в отличие от NSL-KDD, он содержит больше признаков, таких как уровень
энтропии пакетов, флаги TCP и характеристики потока данных. Этот набор данных отличается
сложностью и реалистичностью, что делает его актуальным для современных исследований в
области сетевой безопасности.

Гистограммы позволяют оценить распределение значений числовых признаков. Они
показывают, насколько данные сосредоточены в определенных диапазонах, и помогают
выявить выбросы или дисбаланс в выборке. Например, анализ распределения количества
переданных байтов может показать, что большинство соединений используют небольшие
объемы данных, но есть редкие случаи аномально больших значений, что может указывать на
атаки типа DoS.
Диаграммы рассеяния помогают визуализировать зависимость между двумя числовыми
признаками. Каждая точка на таком графике представляет одно сетевое соединение, а
расположение точек отражает характер взаимосвязи. Если точки образуют четкий тренд, это
может свидетельствовать о линейной зависимости между признаками. Если же они
распределены хаотично, связи может не быть. Диаграммы рассеяния полезны для выявления
кластеров в данных или аномалий, таких как отдельные точки, сильно отличающиеся от
основной массы.
Тепловая карта корреляций используется для анализа взаимосвязей между всеми числовыми
признаками в датасете. Она представляет собой матрицу, в которой значения корреляции между
признаками закодированы цветом. Если два признака имеют высокую положительную
корреляцию, значит, они изменяются вместе, а если отрицательную – увеличение одного
признака сопровождается уменьшением другого. Это помогает отобрать наиболее
информативные признаки и избежать дублирования данных при построении модели.

```python
# Применяем LabelEncoder к столбцу 1
label_encoder = LabelEncoder()
df[1] = label_encoder.fit_transform(df[1])
```

```python
# Повторный вывод данных
print(df.head(100))
```

```python
# Записываем целевую переменную в отдельный фрейм
target_variable = df.iloc[:, 41]

# Удаляем целевую переменную из фрейма
df = df.drop(df.columns[41], axis=1)

# Применяем OneHotEncoder к столбцам 2 и 3
df = pd.get_dummies(df, columns=[2, 3])

# Возвращаем целевую переменную в самый правый край
df['target'] = target_variable.reset_index(drop=True)
```

```python
# Повторный вывод данных
print(df.head(100))
```

```python
# Нормализация числовых признаков
scaler = MinMaxScaler()
df[df.select_dtypes(include=['number']).columns] = scaler.fit_transform(df.select_dtypes(include=['number']))
```

- Oversampling — увеличивает количество примеров меньшинства, например, путем случайного копирования существующих примеров или генерации новых (например, метод SMOTE).
- Undersampling — уменьшает количество примеров большинства, случайным образом удаляя данные, чтобы сбалансировать классы.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Проверка сбалансированности классов
print(df['target'].value_counts())
```

```python
# Преобразование имен всех колонок в строки
df.columns = df.columns.astype(str)

# Разделение данных на признаки (X) и целевая переменная (y)
X = df.drop(columns=['target'])
y = df['target']

# Создание объекта RandomOverSampler
ros = RandomOverSampler(random_state=42)

# Применение oversampling
X_resampled, y_resampled = ros.fit_resample(X, y)

# Создание нового DataFrame с oversampled данными
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)

# Повторная проверка сбалансированности классов
print(df_resampled['target'].value_counts())
```
